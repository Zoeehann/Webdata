```python
import requests
import time

def expand_movie_from_wikidata(qid):
    """
    Fetch related triples for a movie from Wikidata.
    Only fetch attributes of interest to avoid irrelevant data.
    """
    url = "https://query.wikidata.org/sparql"
    
    # Fetch only these attributes: director/actor/genre/production company/release date/language/country
    query = f"""
    SELECT ?p ?o ?oLabel WHERE {{
        VALUES ?p {{
            wdt:P57   wdt:P161  wdt:P136
            wdt:P272  wdt:P577  wdt:P364
            wdt:P495  wdt:P162  wdt:P58
        }}
        wd:{qid} ?p ?o .
        SERVICE wikibase:label {{
            bd:serviceParam wikibase:language "en" .
        }}
    }}
    LIMIT 100
    """
    headers = {
        "User-Agent": "MovieKG-StudentProject/1.0 (han.wang@edu.devinci.fr)",
        "Accept": "application/json"
    }
    try:
        r = requests.get(url, params={"query": query, "format": "json"},
                        headers=headers, timeout=15)
        return r.json()["results"]["bindings"]
    except:
        return []

# Test with Q25188 (Inception)
results = expand_movie_from_wikidata("Q25188")
for row in results[:5]:
    print(row["p"]["value"].split("/")[-1], "|", row["o"]["value"], "|", row.get("oLabel", {}).get("value", ""))
```

```python
from rdflib import Graph, Namespace, URIRef, Literal, RDF, RDFS, OWL
from rdflib import Graph, Namespace, URIRef, Literal, RDF, RDFS, OWL, XSD

MV  = Namespace("http://moviekg.local/resource/")
MVO = Namespace("http://moviekg.local/ontology/")
WD  = Namespace("http://www.wikidata.org/entity/")
WDT = Namespace("http://www.wikidata.org/prop/direct/")

# Load original KB
g = Graph()
g.parse("movie_kb.ttl", format="turtle")
g.parse("alignment.ttl", format="turtle")

print(f"Before expansion: {len(g)} triples")

# Property mapping: Wikidata property → local property
PROP_MAP = {
    "P57":  MVO.hasDirector,
    "P161": MVO.hasActor,
    "P136": MVO.hasGenre,
    "P272": MVO.producedBy,
    "P162": MVO.hasProducer,
    "P58":  MVO.hasScreenwriter,
    "P495": MVO.originCountry,
    "P364": MVO.originalLanguage,
    "P577": MVO.releaseDate,
}

aligned_movies = {}
for local_uri, _, wd_uri in g.triples((None, OWL.sameAs, None)):
    qid = str(wd_uri).split("/")[-1]
    aligned_movies[str(local_uri)] = qid

print(f"Aligned movies: {len(aligned_movies)}")

failed = 0
for i, (local_uri, qid) in enumerate(aligned_movies.items()):
    movie_uri = URIRef(local_uri)
    results = expand_movie_from_wikidata(qid)
    
    for row in results:
        prop_id = row["p"]["value"].split("/")[-1]
        obj_uri = row["o"]["value"]
        obj_label = row.get("oLabel", {}).get("value", "")
        
        if prop_id not in PROP_MAP:
            continue
        
        local_prop = PROP_MAP[prop_id]
        obj_node   = URIRef(obj_uri)
        
        g.add((movie_uri, local_prop, obj_node))
        
        # If label exists, store it as well
        if obj_label:
            g.add((obj_node, RDFS.label, Literal(obj_label, lang="en")))
    
    time.sleep(0.5)
    
    if (i + 1) % 20 == 0:
        print(f"Progress: {i+1}/{len(aligned_movies)}, triples: {len(g)}")

print(f"After expansion: {len(g)} triples")
g.serialize("movie_kb_expanded.ttl", format="turtle")
print("Saved movie_kb_expanded.ttl")
```

```python
# Collect all expanded Person nodes (directors + actors)
persons_to_expand = set()
for s, p, o in g.triples((None, MVO.hasDirector, None)):
    persons_to_expand.add(str(o))
for s, p, o in g.triples((None, MVO.hasActor, None)):
    persons_to_expand.add(str(o))

print(f"Persons to expand: {len(persons_to_expand)}")
```

```python
def expand_person_from_wikidata(qid):
    """Fetch movies a person participated in."""
    url = "https://query.wikidata.org/sparql"
    query = f"""
    SELECT ?movie ?movieLabel WHERE {{
        ?movie wdt:P57 wd:{qid} .
        ?movie wdt:P31 wd:Q11424 .
        SERVICE wikibase:label {{
            bd:serviceParam wikibase:language "en" .
        }}
    }}
    LIMIT 50
    """
    headers = {
        "User-Agent": "MovieKG-StudentProject/1.0 (han.wang@edu.devinci.fr)",
        "Accept": "application/json"
    }
    try:
        r = requests.get(url, params={"query": query, "format": "json"},
                        headers=headers, timeout=15)
        return r.json()["results"]["bindings"]
    except:
        return []

# Expand only wd: prefixed person nodes
wd_persons = [p for p in persons_to_expand if "wikidata.org" in p]
print(f"WD persons: {len(wd_persons)}")
```

```python
for i, person_uri in enumerate(wd_persons):
    qid = person_uri.split("/")[-1]
    person_node = URIRef(person_uri)
    
    results = expand_person_from_wikidata(qid)
    
    for row in results:
        movie_uri   = URIRef(row["movie"]["value"])
        movie_label = row.get("movieLabel", {}).get("value", "")
        
        g.add((movie_uri, RDF.type, MVO.Movie))
        g.add((movie_uri, MVO.hasDirector, person_node))
        if movie_label:
            g.add((movie_uri, RDFS.label, Literal(movie_label, lang="en")))
    
    time.sleep(0.3)
    
    if (i + 1) % 100 == 0:
        print(f"Progress: {i+1}/{len(wd_persons)}, triples: {len(g)}")

print(f"After 2-hop expansion: {len(g)} triples")
g.serialize("movie_kb_expanded.ttl", format="turtle")
```

```python
# Perform batch queries directly using predicates, without relying on existing entities
# Retrieve data in batches using offsets, with 5,000 records per batch, retrieving a total of over 50,000 records
def bulk_fetch_movies(offset=0, limit=5000):
    """Bulk fetch movie triples directly from Wikidata"""
    url = "https://query.wikidata.org/sparql"
    query = f"""
    SELECT ?movie ?director ?actor ?genre WHERE {{
        ?movie wdt:P31 wd:Q11424 .
        ?movie wdt:P57 ?director .
        OPTIONAL {{ ?movie wdt:P161 ?actor . }}
        OPTIONAL {{ ?movie wdt:P136 ?genre . }}
    }}
    LIMIT {limit}
    OFFSET {offset}
    """
    headers = {
    "User-Agent": "Mozilla/5.0 (compatible; StudentProject/1.0)",
    "Accept": "application/sparql-results+json"
    }
    try:
        r = requests.get(url, params={"query": query, "format": "json"},
                        headers=headers, timeout=60)
        return r.json()["results"]["bindings"]
    except Exception as e:
        print(f"Error: {e}")
        return []

# Test once to see how many can be fetched
results = bulk_fetch_movies(offset=0, limit=5000)
print(f"Fetched: {len(results)} rows")
if results:
    print(results[0])
```

```python
print(f"Before bulk: {len(g)} triples")

for offset in range(0, 60000, 5000):
    rows = bulk_fetch_movies(offset=offset, limit=5000)
    
    if not rows:
        print(f"No results at offset {offset}, stopping.")
        break
    
    for row in rows:
        movie_uri    = URIRef(row["movie"]["value"])
        director_uri = URIRef(row["director"]["value"])
        
        g.add((movie_uri,    RDF.type,         MVO.Movie))
        g.add((director_uri, RDF.type,         MVO.Person))
        g.add((movie_uri,    MVO.hasDirector,  director_uri))
        
        if "actor" in row:
            actor_uri = URIRef(row["actor"]["value"])
            g.add((actor_uri, RDF.type,      MVO.Person))
            g.add((movie_uri, MVO.hasActor,  actor_uri))
        
        if "genre" in row:
            genre_uri = URIRef(row["genre"]["value"])
            g.add((genre_uri, RDF.type,      MVO.Genre))
            g.add((movie_uri, MVO.hasGenre,  genre_uri))
    
    print(f"Offset {offset} done, triples: {len(g)}")
    time.sleep(2)

print(f"Final triples: {len(g)}")
g.serialize("movie_kb_expanded.ttl", format="turtle")
print("Saved!")
```

```python
for offset in range(60000, 120000, 5000):
    rows = bulk_fetch_movies(offset=offset, limit=5000)
    
    if not rows:
        print(f"No results at offset {offset}, stopping.")
        break
    
    for row in rows:
        movie_uri    = URIRef(row["movie"]["value"])
        director_uri = URIRef(row["director"]["value"])
        
        g.add((movie_uri,    RDF.type,        MVO.Movie))
        g.add((director_uri, RDF.type,        MVO.Person))
        g.add((movie_uri,    MVO.hasDirector, director_uri))
        
        if "actor" in row:
            actor_uri = URIRef(row["actor"]["value"])
            g.add((actor_uri, RDF.type,     MVO.Person))
            g.add((movie_uri, MVO.hasActor, actor_uri))
        
        if "genre" in row:
            genre_uri = URIRef(row["genre"]["value"])
            g.add((genre_uri, RDF.type,     MVO.Genre))
            g.add((movie_uri, MVO.hasGenre, genre_uri))
    
    print(f"Offset {offset} done, triples: {len(g)}")
    time.sleep(2)
    
    if len(g) >= 60000:
        print("Target reached!")
        break

print(f"Final triples: {len(g)}")
g.serialize("movie_kb_expanded.ttl", format="turtle")
print("Saved!")
```

```python
# Count the number of triples for each relationship
print("Relation statistics")
rel_counts = {}
for s, p, o in g:
    rel_counts[p] = rel_counts.get(p, 0) + 1

for rel, count in sorted(rel_counts.items(), key=lambda x: -x[1]):
    print(f"{str(rel).split('/')[-1]:30s} {count}")

g.serialize("movie_kb_final.ttl", format="turtle")

g.serialize("movie_kb_final.nt", format="nt")

print("\nSaved movie_kb_final.ttl and movie_kb_final.nt")
```
